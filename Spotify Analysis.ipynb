```python
#changing release date to date time format
```

```python
data_tracks.set_index("release_date", inplace =True)
data_tracks.index = pd.to_datetime(data_tracks.index)
data_tracks.head()
```

```python
#filtering specific rows in the dataset
data_tracks[["artists"]].iloc[18]
```

```python
data_tracks['duration'].head()
```

```python
#converting duration from milliseconds to seconds
data_tracks["duration"] = data_tracks["duration_ms"].apply(lambda x: round(x/1000))
data_tracks.drop("duration_ms", inplace = True, axis =1)
data_tracks.duration.head()
```

```python
#correlation map 
```

```python
corr_data = data_tracks.drop(["key","mode","explicit"], axis = 1).corr(method="pearson")
plt.figure(figsize = (15,7))
heatmap = sns.heatmap(corr_data, annot = True,fmt=".1g", vmin=-1,vmax=1,center=0, cmap="inferno", linewidth=1, linecolor="Grey")
heatmap.set_title("Correlation Heatmap Between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_data = data_tracks.sample(int(0.004*len(data_tracks)))
```

```python
print(len(sample_data))
```

```python
#From the correlation map we can see there is a positive relation between loudness and energy. so here we are making regression graph for those two
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_data, y="loudness", x="energy", color = "skyblue").set(title = "Loudness Vs Energy Correlation")
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_data, y="popularity", x="acousticness", color = "orange").set(title = "Popularity Vs acousticness Correlation")
```
